```python
G.in_degree()
```

```python
deg = G.degree()
to_keep = []
for node in G.nodes():
    if deg[node] != 0:
        to_keep.append(node)
    else:
        print("Node: ",node," degree: ",deg[node])
#Create the network only with connected nodes
G=G.subgraph(to_keep)
```

```python
print ("Nodes: ", G.nodes())
```
